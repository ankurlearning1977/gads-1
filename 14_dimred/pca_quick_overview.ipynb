We're going to take a quick look at PCA (Principal Component Analysis), which can be an extremely useful tool for reducing the dimensionality of your dataset.  The basic idea is to look for coordinate axes along which your data varies the most and then simply IGNORE the rest.  These are probably the coordinates in which it would be most fruitful to work with your data, because they are the axes along which the most stuff happens.  They are referred to as the "principal components" - hence the name.

To begin with, let's construct a toy dataset:

```python
import numpy as np
n = 10000 # number of data points
data = np.empty((n,3))
data[:,0:2] = np.random.multivariate_normal([0,0],[[2,0.9],[0.9,0.5]],n)
data[:,2] = np.random.normal(0,1,n)
data -= mean(data,0) # center about mean
data
```

This dataset has three coordinate axes: x0, x1, x2.  The data is highly correlated along the x0 and x1 axes; its behavior along x2 is independent.  Let's take a look at the basic 2-D projections:

```python
fig = plt.figure(); 
# plot x0 vs x1
ax = fig.add_subplot(2,2,1)
ax.set_xlabel('x0'); ax.set_ylabel('x1')
ax.set_xlim(-6,6); ax.set_ylim(-6,6);
ax.set_aspect('equal')
ax.scatter(data[:,0],data[:,1])
# plot x0 vs x2
ax = fig.add_subplot(2,2,2)
ax.set_xlabel('x0'); ax.set_ylabel('x2')
ax.set_xlim(-6,6); ax.set_ylim(-6,6);
ax.set_aspect('equal')
ax.scatter(data[:,0],data[:,2])
# plot x0 vs x1
ax = fig.add_subplot(2,2,3)
ax.set_xlabel('x1'); ax.set_ylabel('x2')
ax.set_xlim(-6,6); ax.set_ylim(-6,6);
ax.set_aspect('equal')
ax.scatter(data[:,1],data[:,2])
```

PCA analysis identifies the coordinate axes along which your data varies the most.  In order to get at this information, we'll need to construct the covariance matrix sigma:

```python
sigma = data.T.dot(data)/n
sigma
```

The covariance matrix has the variances var(x0), var(x1), var(x2) along the diagonal; the off-diagonal entries are the covariances x0-x1, x0-x2, and x1-x2.  Let's check that the diagonal entries are correct:

```python
var(data,0) # column-wise variance
```

Since the covariance matrix is symmetric, we can find a rotated set of coordinate axes q0, q1, q2 in terms of which the covariance matrix is diagonal [spectral theorem] - i.e., the data is uncorrelated when considered among pairs of the new rotated axes.  This is called diagonalization, and the axes which provide this new coordinate system are called eigenvectors of the covariance matrix; the corresponding variances are called eigenvalues.  Essentially, we are factoring our covariance matrix sigma as

sigma = Q * S * Q^T

\begin{equation}
\Sigma = Q \cdot S \cdot Q^T =
\left[
  \begin{array}{cccc}
     |  &  |  &  |  \\
    q_0 & q_1 & q_2 \\
     |  &  |  &  | 
  \end{array}
\right]
\left[
  \begin{array}{ccc}
\textrm{var}(q_0)& 0   & 0 \\
0            & \textrm{var}(q_1) & 0 \\
0 & 0 & \textrm{var}(q_2)
  \end{array}
\right]
\left[
  \begin{array}{cccc}
     -  & q_0^T & - \\
     -  & q_1^T & - \\
     -  & q_2^T & - \\ 
  \end{array}
\right]
\end{equation}

The matrices Q, Q^T effect the rotation of our coordinate axes x0, x1, x2 into the q0, q1, q2 axes.  The matrix S is how our covariance matrix looks when computed in the new coordinate system q.  Our data is uncorrelated across the various qi axes, so the coveriance matrix is diagonal, with the only non-zero entries being the variances (along the diagonal).  We can see the rotated coordinate system in the matrix Q: the columns of Q are the new axes q0, q1, q2.

There are several ways to compute Q and S using numpy.  One is to ask for the eigenvectors qi and eigenvalues var(qi).  This will give you the matrix Q and a list of the diagonal entries of S:

```python
evals, Qe = np.linalg.eig(sigma)
S = diag(evals)
S
```

```python
Qe # the matrix of coordinates qi, as returned by the eigenvalue/eigenvector routine
```

Another way is to compute the singular value decomposition (SVD).  The singular value decomposition is a more general way to factor matrices.   Although not all matrices have a basis of eigenvectors, every matrix has an SVD (even non-square ones!).  For positive symmetric matrices such as the covariance matrix, these two factorizations are in fact the same.  However, as SVD is a more generally useful way to approximate arbitrary matrices, you will often see PCA discussion couched in SVD language ("eigenvalues" becomes "singular values").  So let's do it the SVD way, too:

```python
Q, svals, QT = np.linalg.svd(sigma)
S = diag(svals)
S
```

```python
q0 = Q[:,0:1]; q1 = Q[:,1:2]; q2 = Q[:,2:3];
Q # the matrix of coordinates qi, as returned by the SVD routine
```

These two methods of computing S and Q give essentially the same results, although the SVD routine may have returned the axes qi in a different order (and possibly with some sign flips).  Note that the q0, q1, q2 as returned by SVD are ordered in terms of DECREASING VARIANCE.  This is not an accident; the SVD routine does this in anticipation of our using S and the qi to approximate our data with fewer dimensions.

Let's take a look at the results.  To a good approximation, the new coordinate q1 (middle column of Q) is just a (possibly flipped) version of our old coordinate x2 (the independent one).  The coordinates q0 and q2 are both very nearly contained within the old x0-x1 plane.  

```python
plt.scatter(data[:,0],data[:,1])
plt.plot([-9.1,9.1], [ -4.2,4.2])
plt.plot([-4.2,9.1], [4.2, -9.1])
```

You can see here that the q0 axis is precisely the axis along which our data has the most variation in the x0-x1 plane, and the q2 axis is the one along which our data has the least variation.

It is easy to convert our data from the original coordinate system xi to the rotated axes qi by multiplying each data point by Q.  Let's create a rotated version of our data:

```python
dataQ = data.dot(Q)
dataQ
```

```python
fig = plt.figure(); 
# plot q0 vs q1
ax = fig.add_subplot(2,2,1)
ax.set_xlabel('q0'); ax.set_ylabel('q1')
ax.set_xlim(-6,6); ax.set_ylim(-6,6);
ax.set_aspect('equal')
ax.scatter(dataQ[:,0],dataQ[:,1])
# plot q0 vs q2
ax = fig.add_subplot(2,2,2)
ax.set_xlabel('q0'); ax.set_ylabel('q2')
ax.set_xlim(-6,6); ax.set_ylim(-6,6);
ax.set_aspect('equal')
ax.scatter(dataQ[:,0],dataQ[:,2])
# plot q0 vs q1
ax = fig.add_subplot(2,2,3)
ax.set_xlabel('q1'); ax.set_ylabel('q2')
ax.set_xlim(-6,6); ax.set_ylim(-6,6);
ax.set_aspect('equal')
ax.scatter(dataQ[:,1],dataQ[:,2])
```

You can see clearly from these plots that the data is uncorrelated across the different pairs of q0, q1, and q2 axes.  Also, the data varies most significantly along the q0 axis, followed by q1 and then q2.

To check further that everything works as we expect, let's recompute the covariance matrix in the new coordinate system and check that it agrees with the matrix S that numpy gave us: 

```python
sigmaQ = dataQ.T.dot(dataQ)/n
sigmaQ
```

```python
S # sigmaQ is pretty close!!!
```

Great!  Now: what have we done exactly, and how does it help us simplify our data?  The important point is that we have found a new coordinate system in which each coordinate's worth of data is UNCORRELATED with the others.  In this coordinate system, OUR DATA LOOKS LIKE A BUNCH OF INDEPENDENT VARIABLES.  And, importantly, because we simply rotated our coordinate axes about the mean, WE HAVE PRESERVED THE VARIANCE.  

Let's check that last point.  The total variance of our data, as measured in our original coordinate system, is the same as the variance in our new coordinate system:

```python
diag(sigma)
```

```python
sum(diag(sigma))
```

```python
diag(sigmaQ)
```

```python
sum(diag(sigmaQ))
```

Our data varies so little along the q2 axis, compared to q0 and q1, that if we were to COMPLETELY IGNORE q2 and squint our eyes, we might not notice much had gone missing!  In fact, the q0 and q1 coordinates alone are sufficient to account for almost 98 percent of all the variation in our dataset:

```python
var_q0q1 = sigmaQ[0,0] + sigmaQ[1,1]
var_total = sum(diag(sigmaQ)) 
var_q0q1/var_total
```

This is the key idea of PCA: just ignore the q2 coordinate.  

Our dimensionally-reduced data, then, is simply our dataQ truncated to the first two coordinates q0 and q1 (the principal components):

```python
data_dimred = dataQ[:,0:2]
data_dimred
```

That's really all there is to PCA.  You can now go about your data analysis using the dimensionally-reduced data.  If you haven't thrown away too much information, you should still be able to identify important signals and relationships in your data.  However, because you now have fewer coordinates to work with, the analysis should be easier to perform.

A couple of final notes:

- There is an element of judgement that enters PCA when we select our principal components.  Which of the qi do we keep, and which do we ignore?  There are a couple of popular techniques.  One is to select a percentage threshold (e.g., 95 percent) and keep however many of the qi are required to preserve at least that much of the original variation.  Another is to plot the eigenvalues and look for a "kink" or "elbow" which might signify a significant decrease in the meaningful-ness of the variation arising from the subsequent eigenvectors.

- Sometimes the original data, expressed in terms of the xi, is normalized along each coordinate axis so that the data has approximately equal variance along each xi before performing PCA.  This may or may not be appropriate, depending on what the data represents and what it is that you want PCA to do for you.  Normalization in this manner places all of the original xi on equal footing so far as PCA is concerned.  PCA performed on such a normalized dataset will identify correlations between different coordinates and try to remove redundancy in the data due to correlation.  However, it will of course ignore the absolute magnitude of the variation.  If you wish PCA to take the absolute magnitude of the variation along each axis into account when prioritizing the new coordinate axes qi, then you should not normalize the data (as we did above).
